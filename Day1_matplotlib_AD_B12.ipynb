```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
x=[1,2,3]
y=[1,4,9]
```

```python
plt.plot(x,y)
plt.show()
```

```python
#req: create a sales graph 
    where in our x-axis as sales
    and out y-aixs as revenue
```

```python
plt.xlabel('sales')
plt.ylabel('revenue')
plt.plot(x,y)
plt.show()
```

```python
#req: give the name to the graph as "sales performance"
```

```python
plt.title("sales performance")
plt.xlabel('sales')
plt.ylabel('revenue')
plt.plot(x,y)
plt.show()
```
